```python
import pandas as pd

# Load raw data
df = pd.read_csv(r"C:\Users\vaibh\OneDrive\Desktop\multi_strategy_generator\data\ETH_1H.csv")

# Rename columns to match required format
df = df.rename(columns={
    "Unix Timestamp": "time",
    "Open":           "open",
    "High":           "high",
    "Low":            "low",
    "Close":          "close",
    "Volume":         "volume",
})

# Keep only required columns
df = df[["time", "open", "high", "low", "close", "volume"]]

# Sort ascending by time (raw data is descending)
df = df.sort_values("time").reset_index(drop=True)

# Save
output_path = r"C:\Users\vaibh\OneDrive\Desktop\multi_strategy_generator\data\ETH_1H_formatted.csv"
df.to_csv(output_path, index=True)

print(f"Done — {len(df)} rows saved to {output_path}")
print(df.head())
```

```python
import yfinance as yf
import pandas as pd

ticker = yf.Ticker("BTC-USD")
df = ticker.history(period="2y", interval="1h")  # changed from 1mo to 60d

df = df.reset_index()

df["time"] = df["Datetime"].astype("int64") // 10**6

df = df.rename(columns={
    "Open":   "open",
    "High":   "high",
    "Low":    "low",
    "Close":  "close",
    "Volume": "volume",
})

df = df[["time", "open", "high", "low", "close", "volume"]]

df = df.sort_values("time").reset_index(drop=True)

output_path = r"C:\Users\vaibh\OneDrive\Desktop\multi_strategy_generator\data\btcusd_60m.csv"
df.to_csv(output_path, index=True)

print(f"Done — {len(df)} rows saved to {output_path}")
print(f"Date range: {pd.to_datetime(df['time'].iloc[0], unit='ms')} → {pd.to_datetime(df['time'].iloc[-1], unit='ms')}")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Example data (you can replace with your dataset)
data = [10, 12, 13, 15, 18, 20, 22, 21, 19, 18, 17, 16]

# Convert to pandas series
ts = pd.Series(data)

# Plot ACF
plot_acf(ts, lags=5)
plt.title("Autocorrelation Function (ACF)")
plt.show()

# Plot PACF
plot_pacf(ts, lags=5)
plt.title("Partial Autocorrelation Function (PACF)")
plt.show()
```
